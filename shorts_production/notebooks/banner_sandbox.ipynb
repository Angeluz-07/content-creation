```python
gradient_options = [
    "linear-gradient(to right, #1d2671, #c33764)",  # blue to red(black font not that legible)
    "linear-gradient(to right, #f9d423 0%, #ff4e50 100%)",  # nice orange(black font)  # candidate    
    "linear-gradient(to left, #f7797d, #fbd786, #c6ffdd)",  # mangolike (black font)
    "linear-gradient(-225deg, #FF3CAC 0%, #562B7C 52%, #2B86C5 100%)",  # red to blue(white font) #candidate but not that legible
    "linear-gradient(-45deg, #3f5efb, #fc466b)",  # red to purple(black font)  #candidate
    "linear-gradient(to right, #24243e, #302b63, #0f0c29)",  # business purple(white font, it fades)
    "linear-gradient(to right, #b06ab3, #4568dc)",  # vampiristic style(black font)  #candidate
    "linear-gradient(-45deg, #434343 0%, #000000 100%)",  # black basic gradient linear-gradient(-45deg, #434343 0%, #000000 100%)
]
for color in gradient_options[:3]:
    banner = (
        banner_mod.BasicBanner(
            width=1100, height=320, font_path=font_path, text=text
        )
        .set_background_color(color)
        .render()
    )
    img = banner.get_img()
    display(Image(data=img.encodeToData()))
```
